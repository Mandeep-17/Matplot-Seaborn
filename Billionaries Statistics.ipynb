```python
data=data.drop(columns={'birthDate','lastName','firstName','date','residenceStateRegion','state','birthYear','birthMonth','birthDay','cpi_country',
                      'cpi_change_country','gross_tertiary_education_enrollment','gross_primary_education_enrollment_country','population_country'},axis=1)
```

# Scatter plot of age vs. final worth

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='finalWorth', data=data, hue='category', palette='Set1', alpha=0.7)
plt.title('Age vs. Final Worth',size=15)
plt.xlabel('Age')
plt.ylabel('Final Worth')
plt.legend(title='Category', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.grid()
plt.show()
```

# Bar plot of the top 10 categories

```python
top_categories = data['category'].value_counts().nlargest(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis')
plt.title('Top 10 Categories',size=15)
plt.xlabel('Count')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

# Pie Chart of Industry Distribution

```python
industry_final_worth = data[['industries', 'finalWorth']]
```

```python
industry_totals = industry_final_worth.groupby('industries')['finalWorth'].sum()
```

```python
plt.figure(figsize=(10, 8))
plt.pie(industry_totals, labels=industry_totals.index, autopct='%1.1f%%', startangle=12)
plt.title('Percentage Distribution of Wealth Across Industries',size=15)
plt.show()
```

# Map Showing Locations of Individuals

```python
import geopandas as gpd
import folium
from folium.plugins import MarkerCluster
```

```python
spatial_data = data[['latitude_country','longitude_country','country']]
spatial_data=spatial_data.dropna()

df = spatial_data.copy()
df = df.rename(columns={'latitude_country': 'Latitude',
                        'longitude_country': 'Longitude',
                        'country':'Country'})
```

```python
billionaires_count = df.groupby('Country').size().reset_index(name='Billionaires')
```

```python
billionaires_count
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world = world.merge(billionaires_count, how='left', left_on='name', right_on='Country')
```

```python
world['Billionaires'] = world['Billionaires'].fillna(0)
```

```python
world.head()
```

```python
m = folium.Map(location=[30, 0], zoom_start=2)
```

```python
folium.Choropleth(
    geo_data=world,
    name='choropleth',
    data=billionaires_count,
    columns=['Country', 'Billionaires'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.6,
    line_opacity=0.3,
    legend_name='Number of Billionaires',
).add_to(m)
```

# Average Age comparison of top10 Countries Billionaires to Life Expectancy

```python
d= data[['country','age', 'life_expectancy_country']]
avg_age = d.groupby('country')[['age','life_expectancy_country']].mean().reset_index()
avg_age.columns = ['Country', 'Avg age of billionaires', 'Life Expectancy']
countries = ['United States', 'India','China', 'France','United Kingdom'
            'Russia','Hong Kong', 'Italy','Switzerland','Singapore','Germany']
top_countries_age = avg_age.loc[avg_age['Country'].isin(countries)]
top_countries_age.fillna(85.39)
```

```python
df_long = pd.melt(top_countries_age, id_vars='Country', var_name='Variable', value_name='Value')

plt.figure(figsize=(10,5))

c_plot = sns.barplot(x='Country', y='Value', hue='Variable', data=df_long, palette='Set2')

plt.title('Age comparison of Billionaires w.r.t Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Age')
plt.legend(title='Variable')
    
for p in c_plot.patches:
    c_plot.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```
