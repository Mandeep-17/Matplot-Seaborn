```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the dataset

```python
fifa_data = pd.read_csv('players_20.csv')
fifa_data
```

```python
fifa_data.info
```

```python
fifa_data.isnull().sum()
```

# Histogram of player ages

```python
plt.figure(figsize=(10, 6))
sns.histplot(fifa_data['age'], bins=20, kde=True,color='skyblue')
plt.title('Distribution of Player Ages',size=15)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Scatter plot of height vs weight

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='height_cm',c='red', y='weight_kg', data=fifa_data,alpha=0.6)
plt.title('Scatter Plot of Height vs Weight',size=15)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid()
plt.show()
```

# Bar plot of the top 10 nationalities

```python
top_nationalities = fifa_data['nationality'].value_counts().head(10)
```

```python
plt.figure(figsize=(12, 6))
top_nationalities.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Nationalities of Players',size=15)
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.show()
```


# Box Plot of Player Ages by Nationality

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='nationality', y='age', data=fifa_data.head(100))
plt.title('Box Plot of Player Ages by Nationality',size=15)
plt.xlabel('Nationality')
plt.ylabel('Age')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Nationality vs Clubs

```python
fifa_data['nationality']
```

```python
fifa_data['club']
```

```python
nationality_palette = sns.color_palette("plasma")
club_palette = sns.color_palette("viridis")
```

```python
nationality_count = fifa_data['nationality'].value_counts().head(10)
club_count = fifa_data['club'].value_counts().head(10)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

axes[0].bar(nationality_count.index, nationality_count.values,color=nationality_palette)
axes[0].set_title('Top 10 Nationalities of Players')
axes[0].set_xlabel('Nationality')
axes[0].set_ylabel('Number of Players')
axes[0].tick_params(axis='x', rotation=90)

axes[1].bar(club_count.index, club_count.values,color=club_palette)
axes[1].set_title('Top 10 Clubs with Most Players')
axes[1].set_xlabel('Club')
axes[1].set_ylabel('Number of Players')
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
