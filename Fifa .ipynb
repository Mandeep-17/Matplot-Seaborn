# Player Ages distribution

```python
ages = fifa_data['age'].sort_values()
short_names = fifa_data['short_name'].sort_values()
```

```python
plt.figure(figsize=(10,7))
plt.hist(ages)
plt.title('Player Ages Distribution',size=15)
plt.xlabel('Age')
plt.ylabel('Number of Player')
plt.show()
```

# Height vs Weight

```python
height_cm = fifa_data['height_cm']
```

```python
weight_cm = fifa_data['weight_kg']
```

```python
plt.figure(figsize=(12,7))
plt.scatter(height_cm,weight_cm,color='r',alpha=0.6)
plt.title('Scatter plot of height vs weight',size=20)
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid()
plt.show()
```

# Nationality vs Clubs

```python
fifa_data['nationality']
```

```python
fifa_data['club']
```

```python
nationality_palette = sns.color_palette("plasma")
club_palette = sns.color_palette("viridis")
```

```python
nationality_count = fifa_data['nationality'].value_counts().head(10)
club_count = fifa_data['club'].value_counts().head(10)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

axes[0].bar(nationality_count.index, nationality_count.values,color=nationality_palette)
axes[0].set_title('Top 10 Nationalities of Players')
axes[0].set_xlabel('Nationality')
axes[0].set_ylabel('Number of Players')
axes[0].tick_params(axis='x', rotation=90)

axes[1].bar(club_count.index, club_count.values,color=club_palette)
axes[1].set_title('Top 10 Clubs with Most Players')
axes[1].set_xlabel('Club')
axes[1].set_ylabel('Number of Players')
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
